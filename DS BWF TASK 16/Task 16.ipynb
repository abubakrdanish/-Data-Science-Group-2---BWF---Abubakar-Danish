```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace (0,10,100)
y=np.sin(x)

plt.plot(x,y)
plt.title('Simple Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
y2 = np.cos(x)
plt.plot(x,y,label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Multiple Line Plots')
plt.xlabel('x')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
plt.plot(x,y, marker='o')
plt.title('Line Plot with Markers')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
plt.plot(x, y, linestyle='--', color='r',label='sin(x)')
plt.plot(x, y2, linestyle='-',color='b', label='cos(x)')
plt.title('Customized Line Styles and Colors')
plt.xlabel('x')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y)
plt.title('Basic Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.title('Customized Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x2 = np.random.rand(50)
y2 = np.random.rand(50)

plt.scatter(x, y, c='r', label='Group 1')
plt.scatter(x2, y2, c='b', label='Group 2')
plt.title('Multiple Scatter Plots')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.title("Scatter plot with label and titles")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
categories = ['A', 'B','C','D']
values = [10,24,36,40]
plt.bar(categories, values)
plt.title('Vertical Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
plt.barh(categories,values)
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Category')
plt.show()
```

```python
plt.bar(categories, values, color='skyblue',edgecolor='black')
plt.title('Customized Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes,labels=labels)
plt.title('Basic Pie Chart')
plt.show()
```

```python
colors = ['gold', 'skyblue','orangered','lightcoral']
plt.pie(sizes, labels=labels, colors=colors)
plt.title('Customized Pie Chart')
plt.legend(labels, loc="best")
plt.show()
```

```python
explode = (0,0.1,0,0)
plt.pie(sizes, explode=explode, labels=labels,colors=colors,autopct='%1.1f%%')
plt.title("Explode Pie Chart")
plt.show()
```

```python
plt.pie(sizes, explode=explode, labels=labels,colors=colors,autopct='%1.1f%%')
plt.title("Pie Chart with Percentage Labels")
plt.show()
```
