```python
import pandas as pd

df= pd.read_csv('Road Accident Data.csv')
```

```python
missing = df.isnull().sum()
print(missing)
```

```python
df_drop = df.dropna()
print(df_drop)
```

```python
df_drop_col = df.dropna(axis=1)
print(df_drop_col)
```

```python
df_fill = df.fillna(0)
print(df_fill)
```

```python
df_ffill = df.fillna(method='ffill')
print(df_ffill)
df_bfill = df.fillna(method='bfill')
print(df_bfill)
```

```python
df_interpolate = df.interpolate()
print(df_interpolate)
```

```python
df['Speed_limit'] = df['Speed_limit'].astype(int)
print(df['Speed_limit'].dtype)
```

```python
df['Speed_limit'] = df['Speed_limit'].apply(lambda x:x*1.60934)
print(df['Speed_limit'])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Speed_limit_Normalized'] = scaler.fit_transform(df[['Speed_limit']])
print(df['Speed_limit_Normalized'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Speed_limit_Standardized'] = scaler.fit_transform(df[['Speed_limit']])
print(df['Speed_limit_Standardized'])
```

```python
duplicate_rows = df[df.duplicated()]
print(duplicate_rows)
```

```python
df_no_duplicates = df.drop_duplicates()
print(df_no_duplicates)
```

```python
df_no_duplicates_specific = df.drop_duplicates(subset=['Accident_Index', 'Accident Date'])
print(df_no_duplicates_specific)
```

```python
df['Day_of_Week'] = df['Day_of_Week'].str.lower()
print(df['Day_of_Week'])
```

```python
df['Junction_Control'] = df['Junction_Control'].str.strip()
print(df['Junction_Control'])
```

```python
df['Junction_Control'] = df['Junction_Control'].str.replace('uncontrolled','no control')
print(df['Junction_Control'])
```

```python
df['Accident_Index_Substring']  =df['Accident_Index'].str[:4]
print(df['Accident_Index_Substring'])
```

```python
df['Accident Date'] = pd.to_datetime(df['Accident Date'], errors='coerce', dayfirst=True)
print(df['Accident Date'])
```

```python
df['Year'] = df['Accident Date'].dt.year
df['Month'] = df['Accident Date'].dt.month
df['Day'] = df['Accident Date'].dt.day
print(df[['Year', 'Month', 'Day']])
```

```python
start_date = '2021-01-01'
end_date = '2021-01-31'
df_filtered = df[(df['Accident Date'] >= start_date) & (df['Accident Date'] <= end_date)]
print(df_filtered)
```

```python
df_one_hot = pd.get_dummies(df, columns = ['Day_of_Week'])
print(df_one_hot)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Junction_Control_Label'] = label_encoder.fit_transform(df['Junction_Control'])
print(df['Junction_Control_Label'])
```

```python
severity_mapping = {'Fatal' : 'Severe' , 'Serious' : 'Severe' , 'Slight' : 'Non-Severe'}
df['Accident_Severity'] = df['Accident_Severity'].map(severity_mapping)
print(df['Accident_Severity'])
```

```python
df2 = pd.DataFrame({
    'Accident_Index' : ['200901BS7001' , '200901BS70002'],
    'Accident_Info' : ['Info1' , 'Info2']
})
df_merged = pd.merge(df, df2 , on='Accident_Index',how='left')
print(df_merged)
```

```python
df_concat_vertical = pd.concat([df , df2], axis = 0)
print(df_concat_vertical)
```

```python
df_concat_horizontal = pd.concat([df , df2], axis=1)
print(df_concat_horizontal)
```

```python
df['Total_Vehicles_and_Casualities'] = df['Number_of_Vehicles'] + df['Number_of_Casualties']
print(df['Total_Vehicles_and_Casualities'])
```

```python
df['Speed_limit_bins'] = pd.cut(df['Speed_limit'], bins=3, labels=['Low' , 'Medium', 'High'])
print(df['Speed_limit_bins'])
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
speed_limit_poly = poly.fit_transform(df[['Speed_limit']])
df_poly = pd.DataFrame(speed_limit_poly, columns=['Bias', 'Speed_limit', 'Speed_limit^2'])
df_poly = df_poly.drop(columns=['Bias'])
print(df_poly)
```
