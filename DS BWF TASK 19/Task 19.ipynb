```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")


sns.lmplot(data=titanic, x="age", y="fare", hue="class")
plt.show()
```

```python
sns.residplot(data=titanic, x="age", y="fare")
plt.show()
```

```python
sns.histplot(data=titanic, x="age")
plt.show()
```

```python
sns.kdeplot(data=titanic, x="age")
plt.show()
```

```python
sns.distplot(titanic["age"])
plt.show()
```

```python
sns.boxplot(data=titanic, x="class", y="fare")
plt.show()
```

```python
sns.violinplot(data=titanic, x="class", y="fare")
plt.show()
```

```python
sns.pairplot(data=titanic, hue="survived")
plt.show()
```

```python
numeric_columns = titanic[["age", "fare"]]
correlation_matrix = numeric_columns.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.jointplot(data=titanic, x="age", y="fare", kind="scatter")
plt.show()
```

```python
g = sns.FacetGrid(data=titanic, col="class")
g.map(sns.histplot, "age")
plt.show()
```
